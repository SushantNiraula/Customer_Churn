```python
# Import all the require modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../data/customer_churn_dataset.csv')
```

```python
df.head()
```

```python
# Let's only use few features for this analysis i.e gender, SeniorCitizen, Partner, tenure, PhoneService, Contract, PaymentMethod MonthlyCharges, TotalCharges, Churn
```

```python
X_train=df.drop(['customerID','Dependents','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies'],axis=1)
```

```python
X_train.head()
```

```python
X_train.isnull().sum()
```

```python
X_train['TotalCharges']=pd.to_numeric(X_train['TotalCharges'],errors='coerce')
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
X_train.info()
```

```python
X_train.nunique()
```

```python
trf1=ColumnTransformer([
    ('num',StandardScaler(),['tenure','MonthlyCharges','TotalCharges']),
    ('cat',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[0,2,4,5,6,7,10])
],remainder='passthrough')
```

```python
pipeline=Pipeline([
    ('trf1',trf1)
])
```

```python


# Fit the pipeline
pipeline.fit(X_train)
```

```python
x_train_trf=pipeline.transform(X_train)
```
