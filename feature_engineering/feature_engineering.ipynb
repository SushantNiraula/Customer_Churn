```python
trf1=ColumnTransformer([
    ('binary_encoder',FunctionTransformer(binary_label_encoder),[1,2,3,5,15]),
    ('gender_encoder',FunctionTransformer(gender_encoder),[0]),
    ('one_hot_encoder',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[6,7,8,9,10,11,12,13,14,16]),
    ('binned_columns', KBinsDiscretizer(n_bins=4, encode='onehot-dense', strategy='uniform'), ['tenure','MonthlyCharges'])
],remainder='passthrough')
```

```python
# trf2=SelectKBest(score_func=chi2,k=12)
```

```python
pipe=Pipeline ([
    ('trf1',trf1)
    # ('trf2',trf2)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
X_train_transformed=pipe.transform(X_train)
```
